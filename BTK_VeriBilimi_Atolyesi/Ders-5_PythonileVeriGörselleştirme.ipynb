```python
#Bazen grafiklere sigma, mu, pi vs gibi işaretlerin de
#eklenmesi gerekir. Şimdi bunu nasıl yaptığımıza bakalım
mu1 = 100
sigma1 = 15
x = mu1 + (sigma1 * np.random.randn(10000))

mu2 = 110
sigma2 = 10
y = mu2 + (sigma2 * np.random.randn(10000))

#Söz konusu verilrein histogramı
n, bins, patches = plt.hist(x, 100, density = 1, facecolor="b", alpha=0.25)
n, bins, patches = plt.hist(y, 100, density = 1, facecolor="g", alpha=0.25)

plt.xlabel("Veriler")
plt.ylabel("Olasılıklar")
plt.title(r"Verilerin Histogramı: $\mu_1=100, \ \mu_2=100$")
#r nin anlamı ben bu text içinde bir şeyler yapıcam. Alelade bir başlık değil bu dedik
#$ demek bundan sonra bir ozel bir karakter gelecek demektir ki bu da mü karakteridir.Karakteri yazmadan önce de \ekleriz. _1 ise indisini ekler

plt.annotate("Gauss eğrisinin tepesi", xy=(110,0.04), xytext=(130,0.045),
            arrowprops=dict(facecolor="black", shrink=0.1))
#annotate oku çizer. xy bizim okumuzun ucudur. xytext okun arka kısmını ve textin nerede başlayacağını belirtir

plt.text(75,0.025, r"$\mu_1=100,\ \sigma_1=15$")
plt.text(120,0.035, r"$\mu_2=110,\ \sigma_2=10$")
plt.axis([40, 160, 0, 0.05])
plt.grid(True)
plt.show()
#binomlar, kesirler, iç içe geçmiş kesirler, kökler bile yazılabilir. İNternete bakabilirsin
```
